##### Target variable: Kidnapping outcome

There is a lot of NAs in the target column, since many datapoints do not deal with kidnappings. As can be seen from the .describe(), there should be 14091 observations after a dropna() is run. The codebook says this about this variable: 

This field captures the eventual fate of hostages and kidnap victims. If the “Attack Type” is
“Hostage Taking (Kidnapping),”“Hostage Taking (Barricade Incident),” or a successful
“Hijacking,” then this field applies. The seven values for this field are:

1 = Attempted Rescue <br>
2 = Hostage(s) released by perpetrators<br>
3 = Hostage(s) escaped (not during rescue attempt)<br>
4 = Hostage(s) killed (not during rescue attempt)<br>
5 = Successful Rescue<br>
6 = Combination<br>
7 = Unknown

This means that a possible predictor, "Attack Type" has already been filtered. It is also difficult to tell what category 6 really means, there seems to be a "addnotes" column that could include further details for categories that don't have actual conclusions.

14091 datapoints is a good size for a dataset, though it remains to be seen if there are any nulls for some of the datapoints in the predictors.

### Data Cleaning and Modeling

This sections includes everything pertaining to P4: Modeling and Tuning as well as data cleaning. For section relating to the creation of graphics and other code used in project report, keep going.

```python
#Getting rid of attack type as stated in previous section. It does not provide much value and is difficult to work with.
#This will be revisited if needed.
proj_data = gtd_df[['targtype1','region','nhostkid','nhours','ndays','ransom','weaptype1','hostkidoutcome']]
proj_data = proj_data[proj_data['hostkidoutcome'] >= 0]
```

```python
proj_data = proj_data[proj_data['hostkidoutcome'] != 7]
proj_data['hostkidoutcome'].value_counts()
```

```python
proj_data.shape
```

Decided to drop the 'unknown' category as it is generally inconclusive and it is difficult to determine whether it does provide actual significant value to the analysis. By dropping these datapoints, we lose all observations of kidnappings that do not have a known conclusion (for example, someone gets kidnapped and never seen again, therefore, it is unknown what happened to them). However, this does make the other categories present in the target more prominent and important, which could be good. 

These datapoints could perhaps be used as a sort of extra dataset that the model could try to predict. It would not quite be a validation dataset, as they do not have actual values other than 'unknown' for the target, but if the accuracy of the model created in this project is good enough, these datapoints could be analyzed and categorized based on what the model predicts.

This decision could be changed at a later time depending on results.

```python
#Recoding unknown values for numeric variables as NAs.
proj_data = proj_data.replace(-99.0, np.NaN, regex=True)
proj_data = proj_data.replace(-9.0, np.NaN, regex=True)
```

```python
proj_data.isna().sum()
#As can be seen, NA values did go up, nothing to be too worried about, though nhours and ndays have a lot of NAs still.
```

```python
proj_data['ndays'].value_counts()
```

```python
#Scalar multiplying ndays column in preparation to combine with nhours column
proj_data['ndays'] = proj_data['ndays'].apply(lambda x: x*24)
#This does make the larger values somewhat unreadable, however, this will be binned later, nullifying this issue.
```

```python
#Getting rid of rows that have NA in both nhours and ndays
proj_data = proj_data[(proj_data['nhours'].notna())|(proj_data['ndays'].notna())]
```

```python
proj_data.ndays.fillna(0,inplace=True)
proj_data.nhours.fillna(0,inplace=True)
proj_data['timekid'] = proj_data.ndays + proj_data.nhours
proj_data['timekid'].value_counts()
```

```python
proj_data = proj_data[['targtype1','region','nhostkid','ransom','weaptype1','timekid','hostkidoutcome']]
proj_data
```

```python
proj_data.isna().sum()
```

After combining the nhours and ndays columns that correspond to how long a kidnapping lasted, the dataset is reduced to 4778 observations. This is due to this predictor having a large amount of NA values, which were dropped from the dataset prior to combining. As can be seen from the isna() output, the timekid column (the new name after combination) has no NA values.

```python
proj_data = proj_data.dropna()
```

```python
proj_data['weaptype1'].value_counts()
```

```python
proj_data['targtype1'].value_counts()
```

```python
proj_data = proj_data[(proj_data['weaptype1']!=13) & (proj_data['targtype1']!=20)]
```

```python
proj_data.shape
```

After getting rid of all "unknown" values and NAs in the predictors and target, we are left with ~2800 clean datapoints. This is actually a decent amount, though it certainly could be higher if these "unknown" categories were classified better. It would perhaps be possible to run clustering on text variables to try and categorize these unknown values further, though it is unlikely to yield much if anything.

If any predictors are removed or changed, the amount of data could certainly increase tremendously, this is TBD based on model performance and the impact each variable has. timekid, weaptype1, and ransom are the biggest contributors to the dataset being reduced to this size.

#### Model for category 2

```python
rf_cat2 = RandomForestClassifier(random_state=7).fit(x_train,y_train2)
y_train2.value_counts()
#Baseline accuracy is 50%
```

```python
rf_cat2_cv = cross_val_score(rf_cat2,x_train,y_train2,cv=crossvalidation)
print("Random forest cross validation scores for category 2: ", rf_cat2_cv)
print("Average score:", rf_cat2_cv.mean())
#Model with no parameter tuning is ~65% accurate
```

```python
sns.set(rc={'figure.figsize':(12,8.5)})

importance = pd.DataFrame({'Feature_name': x_train.columns, 
                           'Importance': rf_cat2.feature_importances_})
importance = importance.sort_values(by='Importance', ascending=False)
i_plt = sns.barplot(importance['Importance'], importance['Feature_name'])
i_plt.set(xlabel='Importance (Decrease of Impurity)', ylabel='Feature')
plt.title("Variable Importance for Hostages Released Model (Category 2)")
plt.show()
```

As this predictor list is quite bloated with dummies after one-hot encoding all categoricals, everything other than the largest contributors will be dropped to try to reduce complexity. In this case, the largest 6 contributors will remain (timekid, nhostkid, dumreg_6, dumtarg_14, dumweap_9, ransom). This list of 6 contributors does include a category for each predictor utilized, a larger list that is cutoff at a lower importance level could also be tested depending on results.

```python
x_train_cat2 = x_train[['nhostkid', 'ransom', 'timekid', 'dumtarg_14','dumweap_9','dumreg_6']]
x_test_cat2 = x_test[['nhostkid', 'ransom', 'timekid', 'dumtarg_14','dumweap_9','dumreg_6']]
```

```python
rf_cat2 = RandomForestClassifier().fit(x_train_cat2,y_train2)
rf_cat2_cv = cross_val_score(rf_cat2,x_train_cat2,y_train2,cv=crossvalidation)
print("Random forest cross validation scores: ", rf_cat2_cv)
print("Average score:", rf_cat2_cv.mean())
```

```python
#Parameter tuning
estimators = [1,2,4,8,16,32,64,128,256]
cv_list = []
for estimator in estimators:
    randf = RandomForestClassifier(n_estimators=estimator).fit(x_train_cat2,y_train2)
    randf_cv = cross_val_score(randf, x_train_cat2, y_train2, cv=crossvalidation)
    cv_list.append(randf_cv.mean())
plt.figure(figsize=(16,9))
plt.plot(estimators,cv_list)
plt.show()
```

```python
max_depths = [1,2,4,8,16,32,64,128,256]
cv_list = []
for max_depth in max_depths:
    randf = RandomForestClassifier(max_depth = max_depth).fit(x_train_cat2,y_train2)
    randf_cv = cross_val_score(randf, x_train_cat2, y_train2, cv=crossvalidation)
    cv_list.append(randf_cv.mean())
plt.figure(figsize=(16,9))
plt.plot(max_depths,cv_list)
plt.show()
```

```python
rf_cat2 = RandomForestClassifier(n_estimators=32, max_depth=8).fit(x_train_cat2,y_train2)
rf_cat2_cv = cross_val_score(rf_cat2,x_train_cat2,y_train2,cv=crossvalidation)
print("Random forest cross validation scores: ", rf_cat2_cv)
print("Average score:", rf_cat2_cv.mean())
```

```python
pred = rf_cat2.predict(x_test_cat2)
print(classification_report(y_test2, pred))
print('accuracy is',accuracy_score(pred,y_test2))
print('Confusion Matrix')
print(confusion_matrix(y_test2, pred))
```

After removing complexity and paramter tuning, the final accuracy on the unseen test data is ~64%. This is larger than the baseline accuracy for this model (50%), though not by a large extent. 

```python
#Final feature importance on model
sns.set(rc={'figure.figsize':(12,8.5)})

importance = pd.DataFrame({'Feature_name': x_train_cat2.columns, 
                           'Importance': rf_cat2.feature_importances_})
importance = importance.sort_values(by='Importance', ascending=False)
i_plt = sns.barplot(importance['Importance'], importance['Feature_name'])
i_plt.set(xlabel='Importance (Decrease of Impurity)', ylabel='Feature')
plt.title("Variable Importance for Hostages Released Model (Category 2)")
plt.show()
```

As can be seen, the most important feature, which accounts for a large percentage of the decrease of impurity, is the duration of the kidnapping/hostage incident, this is followed for the number of hostages. This is for the "hostages released" category of the target variable.
